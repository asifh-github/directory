In the analysis of the @dog_rates Twitter account, I was given a CSV file containing the details of the tweets from the account. From the list of tweets, I filtered the tweets that included ratings along with the dates. The ratings in the tweets were given in fractions (10 as the common denominator) and dates as strings. To convert the ratings to numbers, I multiplied them by 10 and converted the dates to a date_time object. Further, I also removed any ratings that were greater than 25 to exclude the obvious biassed ones.

After I had the data in the right format, I was asked if the ratings given by this Twitter account been changing over time? So, I fitted a simple linear regression model, which measures the least squares estimates, with rating as the response variable (y) and date as the explanatory variable (x). 


I found that the slope ($\beta{1}$) of the model is positive, which indicates that the ratings from this Twitter account have been positively changing over time. This also answers our question that: the ratings given by this Twitter account been changing over time. However, to further conclude that the findings are significant, I tested the model for the Null Hypothesis at 95% level of significance, in other words, I looked at the p-value of the model which was largly smaller than 0.05. 

To conduct Null Hypothesis test, three assumptions need to be met. They are:
1. Residuals are independent of each other
2. Residuals are nornally distributed with a mean of zero 
3. Residuals have a constant variance

To check for the assumptions, I plotted a histogram of the residuals, which indicated that the residuals are normally distributed with a mean centred around zero but have some left-skewness. However, we can observe that this skewness occurs due to having lower ratings (outliers) in the years between 2016 and 2017.

Therefore, at this point, I could conclude that the ratings given by this Twitter account have been (increasingly) changing over time at a 95% level of significance. And the model can be used to predict the average dog ratings provided the date.
