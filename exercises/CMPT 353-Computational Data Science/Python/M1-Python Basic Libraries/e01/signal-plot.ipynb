```python
# import munpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# assign no. of samples
N_SAMPLES = 500
```

```python
# create array[samples] for signal 
# one complete sine curve is 2pi
input_range = np.linspace(0, 2*np.pi, N_SAMPLES, dtype=np.float32)
input_range.shape
```

```python
# convert array to simulate signals
signal = np.sin(input_range)
signal
```

```python
# create noise values
mu = 0
sigma = 1
noise = np.random.normal(mu, sigma, N_SAMPLES)
assert noise.shape == input_range.shape
noise.shape
```

```python
# simulate noisy signal
noisy_signal = signal + noise/5
noisy_signal
```

```python
# plot graphs
plt.plot(input_range, noisy_signal, 'b.', alpha=0.5)      # noise vs. time
plt.plot(input_range, signal, 'r-', linewidth=4)          # signal vs. time
plt.legend(['Sensor readings', 'Truth'])
plt.xlabel('Time')
plt.ylabel('Value of thing we care about')
plt.show()
```
