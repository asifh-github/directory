```python
import pandas as pd
```

```python
# read first data
data1 = pd.read_csv('data-1.csv')
data1
```

```python
# mean, standard deviation, and range (min and max) of both the x and y variables
mean_x, mean_y = data1['x'].mean(), data1['y'].mean()
sd_x, sd_y  = data1['x'].std(), data1['y'].std()
min_x , max_x = data1['x'].min(), data1['x'].max()
min_y , max_y = data1['y'].min(), data1['y'].max()
```

```python
# correlation coefficient (r) between the x and y
r = data1['x'].corr(data1['y'])
```

```python
# write function to extract summary
def summary(n):
    # create df with col names
    df = pd.DataFrame(columns=['filename','x_min', 'x_max', 'x_mean', 'x_sd', 'y_min', 'y_max', 'y_mean', 'y_sd', 'r'])
    for i in range(n):
        file = "data-" + str(i+1) +".csv"
        # read data
        data = pd.read_csv(file)
        df.loc[i, 'filename'] = file
        # mean, standard deviation, and range (min and max) of both the x and y variables
        df.loc[i, 'x_min'], df.loc[i, 'x_max'] = data['x'].min(), data['x'].max()
        df.loc[i, 'y_min'], df.loc[i, 'y_max'] = data['y'].min(), data['y'].max()
        df.loc[i, 'x_mean'], df.loc[i, 'y_mean'] = data['x'].mean(), data['y'].mean()
        df.loc[i, 'x_sd'], df.loc[i, 'y_sd'] = data['x'].std(), data['y'].std()
        # correlation coefficient (r) between the x and y
        df.loc[i, 'r'] = data['x'].corr(data['y'])
    
    return df
        
        
    
```

```python
df = summary(6)
df
```
