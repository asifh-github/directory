##### Q4: Pines- 1997 vs. 1990



##### a

```r
# read csv file 
pines_df <- read.csv('ex01-36Pines.csv', header=T)
dim(pines_df)
head(pines_df)
```

```r
# remove na values from df
pines_df = na.omit(pines_df)
dim(pines_df)
head(pines_df)
```

```r
# plot scatter-plot for height of pine trees in 1997 vs. 1990
plot(pines_df$Hgt90, pines_df$Hgt97, col='blue', pch=16, 
     xlab='1990', ylab='1997', main='Scatter plot for the height of trees in 1990 vs. in 1997')
```

Comment: From the above graph, it is hard to determine any relationship between the initial height and the height in 1997.

##### b


```r
# fit a least squares line
# use linear regression 
model <- lm(Hgt97 ~ Hgt90, data=pines_df)
summary(model)
length(pines_df$Hgt90)
length(model$fitted.values)
```

Fitted model: $\hat{Hgt97} = 327.5083 + (1.5257) * Hgt90$

```r
# # plot scatter-plot for height of pine trees in 1997 vs. 1990
plot(pines_df$Hgt90, pines_df$Hgt97, 
     xlab='1990', ylab='1997',
     main='Fitted model for the height of trees in 1990 vs. in 1997',
     pch=16, col='blue', cex=2)
# add fitted model to the scatter plot
lines(pines_df$Hgt90, model$fitted.values, col='red', lwd=3)
```

##### c

```r
plot(model)
```

```r
# plot residual vs fitted value 
plot(model$fitted.values, model$residuals, xlab='Fitted Values', ylab='Residuals', 
     main='Residual vs. Fitted Value Plot', pch=16, cex=1)
abline(h=0, lwd=3, col='red')
```

The plot indicates that there violation in/of the constant variance property. Thus, the errors have non-zero mean.

```r
# plot normal quantile
qqnorm(model$residuals, pch=16)
qqline(model$residuals, col="red", lwd=3)
```

The bow-shaped pattern of the plot indicates the errors have skewness, Therefore, they are not symmetrically distributed.

Explanation: By looking at the above two plots, we can say that the initial height is not a good predictor for predicting the height of trees in 1997.

##### Q6: Textbook prices

##### a

```r
# read csv file 
tbp_df <- read.csv("ex02-16TxtPrc.csv", header=T)
dim(tbp_df)
head(tbp_df)
```

```r
# fit a least squares line
# use linear regression 
model2 <- lm(Price ~ Pages, data=tbp_df)
summary(model2)
```

```r
# # plot scatter-plot of price vs pages
plot(tbp_df$Pages, tbp_df$Price, 
     xlab='Pages', ylab='Price',
     main='Fitted model of Price vs. Pages',
     pch=16, col='blue', cex=2)
# add fitted model to the scatter plot
lines(tbp_df$Pages, model2$fitted.values, col='red', lwd=3)
```

Fitted model: $\hat{Price} = -3.42231 + (0.14733) * Pages$

```r
#plot(model2)
```

##### Significance test

Hypotheses: <br/>
$H_0: \beta_1 = 0$ vs. $H_1: \beta_1 \neq 0$ <br/>
Null hypothesis indicates there is no replationship between y and x and rejection of null indicates that there is a relationship. <br/><br/>


Test statistis: <br/>
t = $\hat{\beta_1}/se(\hat{\beta_1})$ = 7.653 <br/><br/>


P value: <br/>
2.45e-08 < 0.05 (very small, thus, there is really a small chance not to reject null hypothesis) <br/><br/>


Decision: <br/>
We can reject null hypothsis at o.o5 level of significance. <br/>
Therefore, the relationship between Price and pages is significant. <br/><br/> 

##### b

```r
# calculate confidence interval 
confint(model2, "Pages", level=0.95)
```

```r
# use t-table for t
qt(0.975, 28)
```

$\textbf{95% Confidence Interval for Slope}$ <br/><br/>

LL: $\beta_1 - t_{n-2,0.025}*SE(\beta_1)$ = 0.14733 - 2.048 * 0.01925 <br/>
UL: $\beta_1 + t_{n-2,0.025}*SE(\beta_1)$ = 0.14733 + 2.048 * 0.01925 <br/><br/>

Therefore, 95% confidence interval of slope is (0.1078, 0.1867) <br/><br/>

The confidence interval of slope indicates that- we can be 95% confident that the price of a textbook will increase within 0.1078 and 0.1867 if one more page is added to the textbook. It increases becasue the slope of the model is 0.1473 which indicates that there is a positive relationship between price and pages.
