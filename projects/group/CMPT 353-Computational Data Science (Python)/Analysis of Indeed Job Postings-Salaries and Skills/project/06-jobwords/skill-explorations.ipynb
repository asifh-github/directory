```python
import pandas as pd
import nltk
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
for word in ['data','data dcience','Experience','Computer','Ability','work']:
    STOPWORDS.add(word.lower())
```

```python
job_df = pd.read_csv('../03-cleaning/for-anova/anova-data.csv')
job_df['job_key'] = job_df.job_key.str.replace('job-','')
skills_df = pd.read_csv('./skills.csv')

df = job_df.set_index('job_key').join(skills_df.set_index('job_key'))
df = df[['role','seniority','is_intern','skill','scien']]
df = df.loc[~df.skill.str.lower().isin(STOPWORDS)].dropna()
df
```

```python
counts = df.groupby('skill').skill.count().sort_values(ascending=False)
```

```python
wc = WordCloud(background_color='white').generate_from_frequencies(counts)
plt.imshow(wc)
plt.axis('off')
```
